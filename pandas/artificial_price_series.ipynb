# Artificial Prices Series

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Create artificial "prices" using normally distributed returns. **This can result in negative prices**

```python
rng = pd.date_range('1/1/2018', periods=288, freq='H')
df = pd.DataFrame(np.random.normal(0, 1, (len(rng), 2)).cumsum(axis=0), index=rng)
df.head(10)
```

```python
df.plot()
plt.show();
```
