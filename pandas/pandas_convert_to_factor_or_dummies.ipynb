# Convert Vector of Strings to Factor or Dummy in Pandas

```python
import pandas as pd
```

## Create a dataframe

```python
df = pd.DataFrame({'temp': ['hot', 'cold', 'freezing', 'cool', 'warm', 'freezing', 'cold', 'cold', 'hot', 'warm'],
                  'result': ['pass', 'fail', 'fail', 'pass', 'pass', 'fail', 'pass', 'fail', 'pass', 'pass']})
df
```

## As categorical variable, ordinality not preserved

```python
result_factor = pd.factorize(df['result'])
print(result_factor)
df['result_factor'] = result_factor[0]
df
```

## Using dummy variables

```python
pd.get_dummies(df['temp'])
```
