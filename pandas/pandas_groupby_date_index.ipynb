# Groupby Date Index Pandas

```python
import numpy as np
import pandas as pd
```

## Create some sub-daily time series data

```python
rng = pd.date_range('1/1/2018', periods=72, freq='H')
df = pd.DataFrame(np.random.random((len(rng), 5)), index=rng)
df.head(10)
```

## Groupby date

```python
daily_groups = df.groupby(df.index.date)
```

## Output data by day

```python
for idx, day in daily_groups:
    print("Data For {}:\n".format(idx), day)
```
