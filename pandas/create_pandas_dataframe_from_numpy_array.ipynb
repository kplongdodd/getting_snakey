# Create a Pandas DataFrame from a Numpy Array

```python
import numpy as np
import pandas as pd
```

## Create a 2D numpy array

```python
my_array = np.random.random((10, 5))
my_array
```

## Basic DataFrame

```python
df = pd.DataFrame(my_array)
df
```

## DataFrame with named columns

```python
df = pd.DataFrame(my_array, columns=['A', 'B', 'C', 'D', 'E'])
df
```

## DataFrame with named columns and index from a column

```python
df = pd.DataFrame(my_array[:,1:], index=my_array[:, 0], columns=['B', 'C', 'D', 'E'])
df
```
