# Reverse the Order of Rows in a Pandas DataFrame

The last row becomes the first

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame(np.random.random((10, 5)), columns=['A', 'B', 'C', 'D', 'E'])
data
```

## Reverse using `df.sort_index()`

```python
data.sort_index(axis=0 ,ascending=False)
```

## Reverse using `df.iloc[]`

```python
data.iloc[::-1]
```
