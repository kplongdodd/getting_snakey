# Scatterplot Conditioned on a Variable in Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

## Make some data

```python
x = np.random.random(1000)
y = np.random.random(1000)
z = np.random.random(1000)
```

## Create scatterplot of $(x,y)$ conditioned on $z$

Use the `cmap` argument to `matplotlib.pyplot.scatter()`:

http://matplotlib.org/examples/color/colormaps_reference.html

### Grayscale

```python
plt.scatter(x, y, c=z, s=50, cmap='gray');
```

### Sequential colors

```python
plt.scatter(x, y, c=z, s=50, cmap='inferno');
```

### Diverging colors and a colorbar

```python
plt.scatter(x, y, c=z, s=50, cmap='seismic');
plt.colorbar();
```
