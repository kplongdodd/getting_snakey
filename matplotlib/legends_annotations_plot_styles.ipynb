# Legends, Annotations, Plot Styles


```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt

fig_size = (10, 6)
plt.rcParams["figure.figsize"] = fig_size
%config InlineBackend.figure_format = 'retina'
```

```python
y1 = np.array([2, 4, 6, 8, 10, 8, 6, 4, 2])
y2 = np.array([5, 4, 3, 2, 1, 1, 1, 1, 4])
y3 = np.array([0, 0, 1, 1, 1, 2, 2, 2, 2])
y4 = np.array([0, 8, -7, 12, -9, 11, -6, 5, -4])
x = np.array(range(0, len(y1)))
```

## Legends
Use the argement `label` in `plt.plot` to specify labels. Then use `plt.legend` to display the labels in a legend

```python
# Specify the label 'y1'
plt.plot(x, y1, color='red', label='y1') 

# Specify the label 'y2' 
plt.plot(x, y2, color='blue', label='y2')

# Add a legend at the lower center
plt.legend(loc='lower center')

# Add axis labels and title
plt.xlabel('x')
plt.ylabel('Some Quantity')
plt.title('Legend Demonstration')
plt.show()
```

## Annotations

```python
# Plot with legend as before
plt.plot(x, y1, color='red', label='y1') 
plt.legend(loc='lower center')

# Compute the maximum of y1
y1_max = y1.max()

# Calculate the correspondng x value
x_max = x[y1.argmax()]

# Add a black arrow annotation
plt.annotate('Maximum',xy=(x_max, y1_max), xytext=(x_max+0.75, y1_max-0.5), arrowprops=dict(facecolor='black'))

# Add axis labels and title
plt.xlabel('x')
plt.ylabel('Some Quantity')
plt.title('Annotation Demonstration')
plt.show()
```

## Changing Styles

```python
# Import matplotlib.pyplot
import matplotlib.pyplot as plt

# Set the style to 'ggplot'
plt.style.use('ggplot')

# Create a figure with 2x2 subplot layout and make the top left subplot active
plt.subplot(2,2,1)

# Plot in blue y1
plt.plot(x, y1, color='blue')
plt.title('y1')

# Make the top right subplot active in the current 2x2 subplot grid 
plt.subplot(2,2,2)

# Plot in red y2
plt.plot(x, y2, color='red')
plt.title('y2')

# Make the bottom left subplot active in the current 2x2 subplot grid
plt.subplot(2,2,3)

# Plot in green y3
plt.plot(x, y3, color='green')
plt.title('y3')

# Make the bottom right subplot active in the current 2x2 subplot grid
plt.subplot(2,2,4)

# Plot in yellow y4
plt.plot(x, y4, color='yellow')
plt.title('y4')

# Improve the spacing between subplots and display them
plt.tight_layout()
plt.show()
```
