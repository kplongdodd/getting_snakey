# Subplots using Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig_size = (10, 8)
plt.rcParams["figure.figsize"] = fig_size
```

## Make some data

```python
x = [1,2,3,4,5]
y1 = [2,4,6,8,10]
y2 = [500, 400, 300, 200, 100]
y3 = [0, 0, 1, 1, 1]
y4 = [0, 8, -27, 12, -9]
```

## Grid layout

```python
fig = plt.figure()

ax1 = fig.add_subplot(221) # subplot(rows cols subplot_num)
ax1.plot(x, y1, 'r-')

ax2 = fig.add_subplot(222)
ax2.plot(x, y2, 'k-')

ax3 = fig.add_subplot(223)
ax3.plot(x, y3, 'b-')

ax4 = fig.add_subplot(224)
ax4.plot(x, y4, 'g-')


plt.tight_layout()
plt.show()
```

## Share x axis

```python
ax1 = plt.subplot(211)
plt.plot(x, y1)
plt.setp(ax1.get_xticklabels(), fontsize=12)

# share x axis only
ax2 = plt.subplot(212, sharex=ax1)
plt.plot(x, y4)

# make tick labels on lower plot invisible
plt.setp(ax2.get_xticklabels(), visible=False)

plt.show()
```
