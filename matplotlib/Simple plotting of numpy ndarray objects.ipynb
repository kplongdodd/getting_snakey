# Simple plotting of numpy ndarray objects

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Note that we don't need to pass a list of x's as matplotlib can use the indexes of an ndarray!

```python
y = np.random.standard_normal(50)
plt.plot(y);
```

```python
plt.plot(range(len(y)), y);
```

This also works when we pass the return object of an ndarray method. This is because most methods also return an ndarray object:

```python
plt.plot(y.cumsum());
```
