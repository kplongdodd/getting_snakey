# Subplots in a Loop

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
rng = pd.date_range('1/1/2018', periods=24*5, freq='H')
df = pd.DataFrame(np.random.randn(len(rng), 2), index=rng, columns=['Returns', 'Signal'])
display(df.head())
```

```python
%matplotlib inline
return_periods = [1, 2, 3, 4, 5, 6]

fig, axs = plt.subplots(2,3, figsize=(15, 10), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = .5, wspace=.1)
axs = axs.ravel()

for idx, return_period in enumerate(return_periods):

    axs[idx].scatter(df["Signal"], df["Returns"].shift(return_period).rolling(return_period).sum().shift(-return_period), s=15, alpha=0.4)
    axs[idx].set_title("Signal vs {}-Period Future Return".format(return_period))

fig.show();
```
