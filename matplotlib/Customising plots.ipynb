# Customising plots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y = np.random.standard_normal(50)
```

Base plot with no frills

```python
plt.plot(y.cumsum());
```

Add gridlines and restrict axis limits to data

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.axis("tight");
```

Add a little space around the data limits

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.xlim(-1, 50)
plt.ylim(np.min(y.cumsum()) - 1,
np.max(y.cumsum()) + 1);
```

Title, labels, colors, linewidths and dots

```python
plt.figure(figsize=(7, 4)) # adjust figure size
plt.plot(y.cumsum(), "g", lw=2) # change line color and thickness
plt.plot(y.cumsum(), "ro") # add red points
plt.grid(True)
plt.axis("tight")
plt.xlabel("index") # add an x-label
plt.ylabel("value") # add a y-label
plt.title("A Simple Plot"); # add a title
```

Plot a 2D array

```python
y = np.random.standard_normal((50, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(7, 4))
plt.plot(y, lw=1.5) # plot two lines 
plt.plot(y, "k^") # plot markers
plt.grid(True)
plt.axis("tight")
plt.xlabel("index")
plt.ylabel("value")
plt.title("Another Simple Plot");
```
