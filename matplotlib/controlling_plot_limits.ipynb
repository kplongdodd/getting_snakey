# Controlling Plot Limts

Using `xlim`, `ylim` and `axis`

```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt

fig_size = (10, 6)
plt.rcParams["figure.figsize"] = fig_size
%config InlineBackend.figure_format = 'retina'
```

```python
rng = pd.date_range('1/1/2018', periods=288, freq='H')
df = pd.DataFrame(np.random.normal(0, 1, (len(rng), 2)).cumsum(axis=0), columns=['STKA', 'STKB'], index=rng)
df.head(10)
```

```python
# Plot the series STKA and STKB
df.plot()

# Add the axis labels
plt.xlabel('Datetime')
plt.ylabel('Price')

# Set the x-axis range
plt.xlim('2018-01-02 09:00', '2018-01-03')

# Set the y-axis range
plt.ylim(2, 8)

# Add a title and display the plot
plt.title('Price of Stocks A and B')
plt.show()
```

## Use axis() instead of xlim() and ylim()

```python
# Plot the series STKA and STKB
df.plot()

# Add the axis labels
plt.xlabel('Datetime')
plt.ylabel('Price')

# Set the x-axis and y-axis limits
plt.axis(('2018-01-02', '2018-01-03', 2, 8))

# Add a title and display the plot
plt.title('Price of Stocks A and B')

# show the figure
plt.show()
```
